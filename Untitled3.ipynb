```python
class Arbol:
  def __init__(self, valor = None):
    if valor != None:
        self.valor = valor
    else:
        self.valor = None
    self.Izquierda = None
    self.Derecha = None

  def Insertar(self, valor):
    if self.valor:
        if valor < self.valor:
            if self.Izquierda is None:
                self.Izquierda = Arbol(valor)
            else:
                self.Izquierda.Insertar(valor)
        elif valor > self.valor:
            if self.Derecha is None:
              self.Derecha = Arbol(valor)
            else:
              self.Derecha.Insertar(valor)
    else:
        self.valor = valor
        
  def Preorden(self):
    print(self.valor)
    if self.Izquierda:
        self.Izquierda.Preorden()
    if self.Derecha:
        self.Derecha.Preorden()
        
  def Inorden(self):
    if self.Izquierda:
        self.Izquierda.Inorden()
    print(self.valor)
    if self.Derecha:
        self.Derecha.Inorden()
        
  def Postorden(self):
    if self.Izquierda:
        self.Izquierda.Postorden()
    if self.Derecha:
        self.Derecha.Postorden()
    print(self.valor)
```

Para ver si lo hace bien vamos a probar con el ejemplo dado en el sitio web.

```python
arbol = Arbol(20)
arbol.Izquierda = Arbol(18)
arbol.Derecha = Arbol(22)
arbol.Insertar(19)
arbol.Insertar(24)
arbol.Insertar(5)
arbol.Insertar(21)

print("Preorden;")
arbol.Preorden()
print("Inorden;")
arbol.Inorden()
print("Postorden;")
arbol.Postorden()
```
